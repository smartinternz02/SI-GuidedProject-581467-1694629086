```python
q1 = df.sulphates.quantile(0.25)
q3 = df.sulphates.quantile(0.75)
```

```python
print(q1)
print(q3)
```

```python
IQR = q3  - q1
IQR
```

```python
lower_limit = q1 - 1.5*IQR
lower_limit
```

```python
upper_limit = q3 + 1.5*IQR
upper_limit
```

```python
df.median()
```

```python
df['sulphates'] = np.where(df['sulphates']>upper_limit,0.9,df['sulphates'])
```
